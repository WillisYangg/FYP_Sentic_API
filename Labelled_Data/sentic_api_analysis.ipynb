```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
test_df = pd.read_csv('clean_d_tweets.csv')
test_df['tweet'] = test_df['tweet'].str.replace(' ', '%20')
test_df = pd.DataFrame(test_df['tweet'])

test_df.head()
```

```python
sentic_website = 'https://sentic.net/api'
sentic_lang = 'en'
sentic_d_api = 'Cjk3BIibd3pvE0.py'
```

```python
def extract_body(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # body = soup.body.get_text()
            return soup
        else:
            print(f"Failed to fetch {url}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None
```

```python
tweet_list = test_df['tweet'].tolist()

result_dict = {}

for tweet in tweet_list:
    url = sentic_website+'/'+sentic_lang+'/'+sentic_d_api+'?text='+str(tweet)
    body = str(extract_body(url)).strip().replace('[', '').replace('%', '').replace(']', '').replace('\n', '')

    result_dict[tweet] = body

print(result_dict)
```

```python
result_dict
```

```python
data_list = [{'tweet': key, 'sentiment': value} for key, value in result_dict.items()]
```

```python
df = pd.DataFrame(data_list)

df
```

```python
df['sentiment'] = df['sentiment'].astype(int)
df['sentiment'] = np.where(df['sentiment']>0, 'negative', 'positive')

df
```

```python
df['sentiment'].value_counts()
```

```python
df.to_csv('test_d_data.csv')
```
